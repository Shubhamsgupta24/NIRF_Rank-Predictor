<h2>VI . Making ANN model</h2>

```python
class Layer_Dense:
    def __init__(self, n_inputs, n_neurons):
        np.random.seed(0)
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))
    
    def forward(self, inputs):
        self.inputs = inputs
        self.output = np.dot(inputs, self.weights) + self.biases

    def backward(self, dvalues):
        self.dweights = np.dot(self.inputs.T, dvalues)
        self.dbiases = np.sum(dvalues, axis=0, keepdims=True)
        self.dinputs = np.dot(dvalues, self.weights.T)

class Activation_ReLU:
    def forward(self, inputs):
        self.inputs = inputs
        self.output = np.maximum(0, inputs)

    def backward(self, dvalues):
        self.dinputs = dvalues.copy()
        self.dinputs[self.inputs <= 0] = 0

class Optimizer_SGD:
    def __init__(self, learning_rate=0.01):
        self.learning_rate = learning_rate
    
    def update_params(self, layer):
        layer.weights += -self.learning_rate * layer.dweights
        layer.biases += -self.learning_rate * layer.dbiases

class Loss:
    def calculate(self, output, y):
        sample_losses = self.forward(output, y)
        data_loss = np.mean(sample_losses)
        return data_loss

class Loss_MeanSquaredError(Loss):
    def forward(self, y_pred, y_true):
        sample_losses = np.mean((y_true - y_pred)**2, axis=-1)
        return sample_losses

    def backward(self, dvalues, y_true):
        samples = len(dvalues)
        outputs = len(dvalues[0])
        self.dinputs = -2 * (y_true - dvalues) / outputs
        self.dinputs = self.dinputs / samples
```

```python
class NeuralNetwork:
    def __init__(self, features, target, hidden_neurons, learning_rate, epochs):
        self.features = features
        self.target = target
        self.hidden_neurons = hidden_neurons
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.model = None

    def normalize_features(self, X):
        # Normalize features to have zero mean and unit variance
        mean = np.mean(X, axis=0)
        std = np.std(X, axis=0)
        
        # Add a small epsilon to avoid division by zero
        epsilon = 1e-8
        X_normalized = (X - mean) / (std + epsilon)
        return X_normalized

    def train(self, X, y):
        X_normalized = self.normalize_features(X)

        layer1 = Layer_Dense(len(self.features), self.hidden_neurons)
        activation1 = Activation_ReLU()
        layer2 = Layer_Dense(self.hidden_neurons, 1)
        optimizer = Optimizer_SGD(learning_rate=self.learning_rate)
        loss_function = Loss_MeanSquaredError()

        for epoch in range(self.epochs):
            layer1.forward(X_normalized)
            activation1.forward(layer1.output)
            layer2.forward(activation1.output)

            loss = loss_function.calculate(layer2.output, y)

            loss_function.backward(layer2.output, y)
            layer2.backward(loss_function.dinputs)
            activation1.backward(layer2.dinputs)
            layer1.backward(activation1.dinputs)

            optimizer.update_params(layer1)
            optimizer.update_params(layer2)

            if epoch % 100 == 0:
                print(f'Epoch {epoch}, Loss: {loss}')

        self.model = layer2
        self.X_normalized = X_normalized

    def predict(self, new_data):
        new_X = np.array([new_data[feature] for feature in self.features]).reshape(1, -1)
    
        # Normalize the new data using the mean and standard deviation from the training set
        new_X_normalized = (new_X - np.mean(self.X_normalized, axis=0)) / (np.std(self.X_normalized, axis=0) + 1e-8)
    
        layer1 = Layer_Dense(len(self.features), self.hidden_neurons)
        activation1 = Activation_ReLU()
    
        layer1.forward(new_X_normalized)
        activation1.forward(layer1.output)
        self.model.forward(activation1.output)
    
        # Scale the output to the desired range (0 to 100)
        prediction = np.clip(self.model.output[0, 0], 0, 100)
    
        return prediction


# Example usage
# Assuming df is your DataFrame and new_data is the new data for prediction
features = ['TLR', 'RPC', 'GO', 'OI', 'Perception']
target = 'Score'
nn = NeuralNetwork(features, target, hidden_neurons= 200 , learning_rate=0.0001, epochs=200000)

# Assuming df is your DataFrame
X = df[features].values
y = df[target].values.reshape(-1, 1)

nn.train(X, y)

new_data = {'TLR': 90, 'RPC': 90, 'GO': 90, 'OI': 90, 'Perception': 90}  # Replace ... with actual values
prediction = nn.predict(new_data)
print("Predicted Score:", prediction)
```
